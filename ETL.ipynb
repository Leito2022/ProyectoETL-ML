```python
#Librerias
import pandas as pd
import numpy as np
```

```python
#Carga de datasets desde el github

dfHulu = pd.read_csv("https://raw.githubusercontent.com/Leito2022/ProyectoETL-ML/main/Datasets/hulu_titles.csv")
dfNet = pd.read_csv("https://raw.githubusercontent.com/Leito2022/ProyectoETL-ML/main/Datasets/netflix_titles.csv")
dfAmz = pd.read_csv("https://raw.githubusercontent.com/Leito2022/ProyectoETL-ML/main/Datasets/amazon_prime_titles.csv")
dfDny =pd.read_csv("https://raw.githubusercontent.com/Leito2022/ProyectoETL-ML/main/Datasets/disney_plus_titles.csv")
```

```python
dfHulu.head()
```

```python
dfHulu["id_pelicula"] = "h" + dfHulu["show_id"]
dfNet["id_pelicula"] = "n" + dfHulu["show_id"]
dfAmz["id_pelicula"] = "a" + dfHulu["show_id"]
dfDny["id_pelicula"] = "d" + dfHulu["show_id"]
```

```python
dfPeli = pd.concat([dfHulu,dfNet,dfAmz,dfDny]).reset_index(drop=True)
```

```python
dfPeli["rating"].isna().sum()
```

```python
dfPeli["rating"].fillna("G",inplace = True)
```

```python
#Sacar espacios en una columna datatime

dfPeli["date_added"] = dfPeli["date_added"].str.lstrip()
```

```python
dfPeli["date_added"] = pd.to_datetime(dfPeli["date_added"], format = "%B %d, %Y")
```

```python
dfPeli['date_added'] = dfPeli['date_added'].dt.strftime('%Y-%m-%d')
```

```python
dfPeli.head()
```

```python
dfPeli = dfPeli.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
dfPeli.sample(10)
```

```python
dfPeli['duration_int'] = dfPeli['duration'].str.extract('(\d+)')
```

```python
def tipo_duracion(columna):
  if columna == "movie":
    return "min"
  else:
    return "season"
```

```python
dfPeli['duration_type'] = dfPeli["type"].apply(tipo_duracion)
```

```python
dfPeli.drop("duration", axis= 1)
```

```python
dfPeli["indice"] = dfPeli.index
```

```python
import pymysql
from sqlalchemy import create_engine
```

```python
# Se establece la conexion con MySQL
cadena_conexion = 'mysql+pymysql://root:admin@localhost:3306/streaming'
conexion = create_engine(cadena_conexion)
```

```python
#Pasar a DB en mysql
dfPeli.to_sql(name = 'servicio_streaming', con = conexion)
```
